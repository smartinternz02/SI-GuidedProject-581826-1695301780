```python
import pandas as pd
import io

df = pd.read_csv(io.BytesIO(uploaded['House Price India.csv']))
print(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
#Univariate Analysis:
#Example: Create a histogram for the 'number of bedrooms' column.
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='number of bedrooms', bins=15, kde=True)
plt.title('Histogram of Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Frequency')
plt.show()
```

```python
#Bivariate Analysis:
#Example: Create a scatter plot for 'living area' vs 'Price'.
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='living area', y='Price')
plt.title('Scatter Plot of Living Area vs Price')
plt.xlabel('Living Area')
plt.ylabel('Price')
plt.show()
```

```python
#Multivariate Analysis:
#Example: Create a pairplot for selected numeric variables.
sns.pairplot(data=df[['living area', 'Price', 'number of bedrooms', 'number of bathrooms']], diag_kind='kde')
plt.suptitle('Pairplot of Living Area, Price, Bedrooms, and Bathrooms')
plt.show()
```

```python
#Descriptive Statistics:
#Example: Calculate and display descriptive statistics for numeric variables.
desc_stats = df[['living area', 'Price']].describe()
print(desc_stats)
```

```python
#Handle Missing Values:
#Example: Impute missing values in the 'Renovation Year' column with the median.
median_renovation_year = df['Renovation Year'].median()
df['Renovation Year'].fillna(median_renovation_year, inplace=True)
```

```python
#Verify that missing values are handled:
missing_values_count = df.isnull().sum()
print("Missing Values Count:\n", missing_values_count)
```
